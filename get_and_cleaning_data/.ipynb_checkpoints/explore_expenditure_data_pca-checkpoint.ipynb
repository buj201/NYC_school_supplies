# Using PCA to explore the school budget data

In this notebook, we use PCA to explore NYC School expenditures.

```python
import pandas as pd

data_2006 = pd.read_csv('../../data/clean_expenditure_data/clean_year_2006.csv', index_col=0)
data_2007 = pd.read_csv('../../data/clean_expenditure_data/clean_year_2007.csv', index_col=0)
data_2008 = pd.read_csv('../../data/clean_expenditure_data/clean_year_2008.csv', index_col=0)
data_2009 = pd.read_csv('../../data/clean_expenditure_data/clean_year_2009.csv', index_col=0)
data_2010 = pd.read_csv('../../data/clean_expenditure_data/clean_year_2010.csv', index_col=0)
data_2011 = pd.read_csv('../../data/clean_expenditure_data/clean_year_2011.csv', index_col=0)
data_2012 = pd.read_csv('../../data/clean_expenditure_data/clean_year_2012.csv', index_col=0)
```

```python
all_years_data = {2006: data_2006, 2007: data_2007, 2008: data_2008, 2009: data_2009, 2010: data_2010, 2011: data_2011, 2012: data_2012}
for year in all_years_data:
    print all_years_data[year].shape
```

We will fill in missing values with 0. This is justified under the assumption that if a row is missing from the expenditure report, it is because the school did not have any expenditures under that category.

```python
for year in all_years_data:
    all_years_data[year].fillna(0,inplace=True)
    print '\nYear = ', str(year)
    print 'Number observations', len(all_years_data[year])
    print 'Number missing', all_years_data[year].isnull().sum().sum()
```

Great! Now let's center and scale all the continuous budget variables.

```python
for year in all_years_data:
    for feature in all_years_data[year].drop(['Title_1', 'District','School'], axis=1):
        ##Center- subtract off mean:
        all_years_data[year][feature] = all_years_data[year][feature] - all_years_data[year][feature].mean()
        ##Scale- divide by max:
        all_years_data[year][feature] = all_years_data[year][feature]/all_years_data[year][feature].max()
```

Now, since a number of features have 0 variance, we can anticipate having introduced NaN values (when dividing by zero- since removing the mean centers all these zero variance variables as a zero vector): let's find them:

```python
for year in all_years_data:
    print '\nYear = ', str(year)
    print 'Number observations', len(all_years_data[year])
    print 'Missing variable counts:'   
    print all_years_data[year].isnull().sum()[all_years_data[year].isnull().sum()>0]
```

Now let's drop them:

```python
for year in all_years_data:
    all_years_data[year].dropna(axis=1, inplace=True)
    print '\nYear = ', str(year)
    print 'Number observations:', len(all_years_data[year])
    print 'Number missing values:', all_years_data[year].isnull().sum().sum()
```

Now let's use PCA and project onto the first two components:

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(5,15))
fig.subplots_adjust(hspace=.75)
for year in all_years_data:
    X = all_years_data[year].drop(['Title_1', 'District','School'], axis=1)
    pca = PCA()
    pca.fit_transform(X)
    ax = plt.subplot(7,1,year-2005)
    ax.bar(np.arange(1,len(pca.explained_variance_ratio_)+1),pca.explained_variance_ratio_)
    ax.set_title(str(year))
    ax.set_ylabel('% variance explained')
    ax.set_yticks(np.arange(0,0.6,0.1), minor=False)
    ax.set_xlabel('Principle component')
```

```python
fig = plt.figure(figsize=(10,70))
for year in all_years_data:
    X = all_years_data[year].drop(['Title_1', 'District','School'], axis=1)
    print X.shape
    pca = PCA()
    trans_x = pca.fit_transform(X)
    ax = plt.subplot(7,1,year-2005)
    mask = (all_years_data[year]['Title_1'] == 'Yes')
    ax.scatter(trans_x[mask.values,0],trans_x[mask.values,1], color='b', label='Title 1 schools')
    ax.scatter(trans_x[~mask.values,0],trans_x[~mask.values,1], color='r', label='Non-Title 1 schools')
    Title = 'Comparing Title 1 and non-Title 1 schools for year = ' + str(year)
    ax.legend(scatterpoints=1)
    ax.set_title(Title)
    ax.set_ylabel('Component 2')
    ax.set_xlabel('Component 1')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for year in range(2006,2013):
    plt.figure(figsize=(8,16))
    ax = plt.subplot(7,1,year-2005)
    ax.hist(all_years_data[year]['Total'],bins=100, range=(-1,1))
    key = 'Median = ' + str(all_years_data[year]['Total'].median())
    ax.axvline(all_years_data[year]['Total'].median(), color='k', linewidth=4, label=key)
    Title = 'Total per pupil for year = ' + str(year)
    ax.set_title(Title)
    ax.set_xlim(-1,1)
    ax.legend()
```
